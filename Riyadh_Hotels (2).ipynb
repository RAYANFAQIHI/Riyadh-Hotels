```python
df = df.dropna(subset=["price"])
```

```python
df['Info'] = df['Info'].fillna('No Info')
```

```python
df.isnull().sum()
```

```python
pd.to_datetime(df['checkIn'])
```

```python
df['checkIn'] = pd.to_datetime(df['checkIn'])
df['checkOut'] = pd.to_datetime(df['checkOut'])
```

```python
df.info()
```

```python
df['stay_days'] = (df['checkOut'] - df['checkIn']).dt.days
df
```

```python
df['Free breakfast'] = df['Info'].str.contains('Free breakfast available', na=False)
df['Free cancellation'] = df['Info'].str.contains('Free cancellation', na=False)
df
```

```python
df = df.drop(columns=['Info'])
df
```

```python
df[['price', 'base_price']].isnull().sum()
```

```python
df[['price', 'base_price']].info()
```

```python
df[['price', 'base_price']].describe().T
```

```python
df['discount'] = df['price'] - df['base_price']
df['discount_percent'] = ((df['price'] - df['base_price']) / df['base_price']) * 100
```

```python
df = df.drop_duplicates()
```
